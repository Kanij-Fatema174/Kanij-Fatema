```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define constants and thresholds

# Define constants and thresholds
average_price = 600  # Average price of the smartphone in BDT
discount_threshold = 0.2  # 20% discount threshold
critical_stock_level = 10  # Critical stock level
default_order_quantity = 10  # Quantity to order when stock is critically low
additional_order_quantity = 15  # Quantity to order during a deal
```

```python
# Initialize the inventory and price data

# Initialize the inventory and price data
stock_levels = [20]  # Starting with 20 units in stock
prices = [600, 580, 550, 530, 500, 520, 540, 610]  # Example price fluctuations
orders = []  # To keep track of orders
```

```python
# prompt: Defining the Decision Logic

# Define the decision logic function
def make_decision(current_stock, current_price):
    order_quantity = 0
    if current_stock <= critical_stock_level:
        order_quantity = default_order_quantity
        if current_price < average_price * (1 - discount_threshold):
            order_quantity += additional_order_quantity
    return order_quantity
```

```python
# prompt: simulating the Agent's Decisions

# Simulate the agent's decisions over time
for i in range(len(prices) - 1):
    current_stock = stock_levels[-1]
    current_price = prices[i]
    order_quantity = make_decision(current_stock, current_price)
    orders.append(order_quantity)
    next_stock = current_stock + order_quantity - np.random.randint(5, 15)  # Simulate sales
    stock_levels.append(max(0, next_stock))  # Stock cannot be negative


# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(stock_levels, label='Stock Level')
plt.plot(prices[:-1], label='Price', linestyle='--')
plt.bar(range(len(orders)), orders, label='Orders', alpha=0.5)

plt.xlabel('Time Step')
plt.ylabel('Value')
plt.title('Inventory Management Simulation')
plt.legend()
plt.grid(True)
plt.show()
```
